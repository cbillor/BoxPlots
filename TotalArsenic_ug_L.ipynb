```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#read COCData.csv data as a pandas dataframe:
df=pd.read_csv('/home/ubuntu/catherine_work/RDM/COCData.csv')
```

```python
#filter df into groups of wells for plotting

#upgradient
wells = ['MW31', 'MW58', 'MW58']
df1 = df[df.WellID.isin(wells)]

#footprint
wells2 = ['MW50', 'MW51', 'MW53', 'MW54', 'MW57', 'MW59']
df2 = df[df.WellID.isin(wells2)]

#downgradient - near
wells3 = ['MW56','MW45', 'MW46', 'MW47', 'MW52', 'MW55','MW44']
df3 = df[df.WellID.isin(wells3)]

#downgradient - far
wells4 = ['MW48', 'MW40', 'MW42', 'MW43', 'MW29']
df4 = df[df.WellID.isin(wells4)]

#MPA
wells5 = ['MW16', 'MW17', 'MW01', 'MW10', 'MW22', 'Mw32']
df5 = df[df.WellID.isin(wells5)]
```

```python
#create boxplot using seaborn
ax=sns.boxplot(x='WellID', y='TotalArsenic_ug_L', data=df1)
ax=sns.swarmplot(x='WellID', y='TotalArsenic_ug_L', data=df1, color="black").set_title('Upgradient Wells')
```

```python
sns.boxplot(x='WellID', y='TotalArsenic_ug_L', data=df2)
ax=sns.swarmplot(x='WellID', y='TotalArsenic_ug_L', data=df2, color="black").set_title('Footprint Wells')
```

```python
sns.boxplot(x='WellID', y='TotalArsenic_ug_L', data=df3)
ax=sns.swarmplot(x='WellID', y='TotalArsenic_ug_L', data=df3, color="black").set_title('Near Downgradient Wells')
```

```python
sns.boxplot(x='WellID', y='TotalArsenic_ug_L', data=df4).set_title('Far Downgradient Wells')
ax=sns.swarmplot(x='WellID', y='TotalArsenic_ug_L', data=df4, color="black")
```

```python
sns.boxplot(x='WellID', y='TotalArsenic_ug_L', data=df5).set_title('MPA Wells to Swap')
ax=sns.swarmplot(x='WellID', y='TotalArsenic_ug_L', data=df5, color="black")
```
